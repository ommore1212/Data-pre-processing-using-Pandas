#Data Normalization

```python
dataset.head()
```

Simple feature Scaling



```python
dataset['horsepower']=dataset['horsepower']/dataset['horsepower'].max()
```

```python
dataset.head()
```

Min - Max

```python
dataset['weight'] = (dataset['weight'] - dataset['weight'].min())/(dataset['weight'].max() - dataset['weight'].min() )
```

```python
dataset.head()
```

Z score


```python
dataset['displacement'] = (dataset['displacement'] - dataset['displacement'].mean()) / dataset['displacement'].std()
```

```python
dataset.head()
```

```python
dataset['origin'].value_counts()
```

```python
dataset.head()
```

#Categorial -> Numerical

```python
encode_data = pd.get_dummies(dataset['origin'])
```

```python
dataset = pd.concat([dataset,encode_data],axis=1)
```

```python
dataset.head()
```

```python
dataset.drop('origin',axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset['europe'] = dataset['europe'].astype(int)
```

```python
dataset['japan'] = dataset['japan'].astype(int)
```

```python
dataset['usa'] = dataset['usa'].astype(int)
```

```python
dataset['model_year'].value_counts()
```

# Binning


```python
bin = np.linspace(min(dataset['model_year']),max(dataset['model_year']),3)
```

```python
group_names = ['Old',"New"]
```

```python
dataset['model_year_binned'] = pd.cut(dataset['model_year'],bin,labels=group_names,include_lowest=True)
```
